```python
# Filtros/Recortes
devsRdropna = devsRdropna[devsRdropna["Country"] == "United States"]
devsRdropna = devsRdropna[devsRdropna["MainBranch"] == "I am a developer by profession"]
#devsRdropna = devsRdropna[devsRdropna["OrgSize"] == "20 to 99 employees"]

devsRdropna
```

```python
devsLimit = devsRdropna[["Respondent", "LanguageWorkedWith", "CompTotal", "CompFreq", "ConvertedComp", "YearsCode", "WorkWeekHrs"]]
devsLimit = devsLimit.dropna()
```

## Grafo

### a) grafo dirigido

```python
h = Graph(directed=False) # dirigido
```

### b) Vértices e propriedades

```python
dev_filtered = pd.DataFrame([devsLimit.Respondent, devsLimit.LanguageWorkedWith]).transpose()
dev_filtered['LanguageWorkedWith'] = dev_filtered['LanguageWorkedWith'].str.split(';')
dev_filtered['Respondent'] = dev_filtered['Respondent'].astype(float).astype(int)
dict_dev = {}
i = 0
while i < len(dev_filtered) - 1:
    dict_dev[i] = dev_filtered.iloc[i, :]
    i+=1
```

```python
dict_dev
```

```python
for dev in dev_filtered['Respondent']:
    h.add_vertex(dev) # Adicionando propriedades
    u = h.get_vertex(dev)
```

```python
print('número de vértices de g:', h.num_vertices())
```

### c) Arestas e propriedades

```python
dict_dev
```

```python
dev_filtered['Respondent'] = dev_filtered['Respondent'].astype('int')
```

```python
#print(dev_filtered)
dict_common = {}
i = 0
language_df = dev_filtered['LanguageWorkedWith']
for key, value in dict_dev.items():
    #print(f'Key: {key}\nValue: {value}')
    while i < len(dict_dev):
        #print(f'i: {i}\nkey: {key}')
        if np.any(pd.isna(language_df.iloc[key])) or np.any(pd.isna(language_df.iloc[i])):
            pass
        else:
            #if value not in dict_common.keys():
            if dict_dev[key]['Respondent'] != dict_dev[i]['Respondent']:
                set_common = set(language_df.iloc[key]).intersection(set(language_df.iloc[i]))
                set_common = tuple(set_common)
                #print(f'Tipo do set common é {type(set_common)}\nSet common é {set_common}')
                tuple_respondent = (dict_dev[key]['Respondent'], dict_dev[i]['Respondent'])
                #print(f'Type do tuple_respondent: {type(tuple_respondent)}')
                tuple_respondent_inverted = (dict_dev[i]['Respondent'], dict_dev[key]['Respondent'])
                if tuple_respondent_inverted not in dict_common.keys():
                    dict_common[tuple_respondent] = set_common
                    #print(dict_common.keys())
                    #print(dict_common.values())
        i+=1
    i=0
```

```python
threshold = 1n 
i = 0
for key, value in dict_common.items():
    if len(value) >= threshold:
        i+=1
        h.add_edge(key[0], key[1])
        #print(f"Cria um aresta entre {key[0]} e {key[1]}")
```
